```python
np.random.seed(44)

t = np.arange(0, 30, 0.01)
nse1 = np.random.randn(len(t))
nse2 = np.random.randn(len(t))

s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1, layout='constrained')
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, NFFT=256, Fs=1. / 0.01)
axs[1].set_ylabel('Coherence')

plt.show()
```

```python
import random
def spinnerA1(probabilityDistribution):
    cumulativeDistribution = []
    cumulativeDistributionSum = 0
    for p in probabilityDistribution:
        cumulativeDistributionSum += p
        cumulativeDistribution.append(cumulativeDistributionSum)
    randomInteger = random.random()
    index = 0
    for p in cumulativeDistribution:
        if randomInteger < p:
            return index
        index += 1
```

```python
spinnerA1([0.2, 0.5, 0.1, 0.2])
```

```python
import numpy as np
def spinnerA2(probabilityDistribution):
    return np.random.choice(len(probabilityDistribution), p=probabilityDistribution)
```

```python
spinnerA2([0.2, 0.5, 0.1, 0.2])
```

```python
def spinnerB(listNumber):
    probabilityDistribution = []
    totalAngle = sum(listNumber)
    for angle in listNumber:
        probabilityDistribution.append(round(angle/totalAngle, 4))
    return spinnerA1(probabilityDistribution)
```

```python
spinnerB([45, 85, 70, 150])
```

```python
import random
import math
import numpy as np
def montCarl(n):
    x_positions = np.random.uniform(-50, 50, n)
    y_positions = np.random.uniform(-50, 50, n)
    distances = np.sqrt(x_positions**2 + y_positions**2)
    validPositions = np.sum(distances <= 29)
    return validPositions/n
```

```python
montCarl(5000000)
```
